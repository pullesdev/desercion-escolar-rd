<h2 style="color:#A23D72; font-family:Georgia;">
 1. Cargar Datos
</h2>

```python
import pandas as pd

df = pd.read_csv(
    '../data/raw/centros_educativos_de_rd_2023_2024.csv',
    encoding='latin-1',
    sep=';'
)
```

<h2 style="color:#A23B72; font-family:Georgia;">
 2. Limpieza de Datos
</h2>

```python
df.columns = df.columns.str.lower()
columnas = df.columns
columnas
```

```python
df.loc[df["regional"] == "18 - BAORUCO", "regional"] = "18 - BAHORUCO"
df.loc[df["sector"] == "PÚBLICO", "sector"] = "PUBLICO"
df.loc[df["provincia"] == "BAORUCO", "provincia"] = "BAHORUCO"
```

```python
# Ver cuántos hay
df["nivel"].value_counts(dropna=False)

# Inspeccionar filas
df[df["nivel"] == "#REF!"]

# Reemplazar por NaN
import numpy as np

df.loc[df["nivel"] == "#REF!", "nivel"] = np.nan
```

```python
df = df.dropna(subset=['coordenadas latitud', 'coordenadas longitud', 'nivel'])
```

```python
df.isnull().sum()
```

```python
df['coordenadas latitud'] = pd.to_numeric(df['coordenadas latitud'], errors='coerce')
df['coordenadas longitud'] = pd.to_numeric(df['coordenadas longitud'], errors='coerce')
```

```python
df.to_csv('../data/processed/centros_educativos_limpios.csv', index=False)
print("Datos limpios guardados correctamente.")
```
